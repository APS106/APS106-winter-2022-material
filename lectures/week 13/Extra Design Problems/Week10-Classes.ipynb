## Tic-Tac-Toe with Python

Using Python, we can create a tic-tac-toe board using ```classes```

We will be using nested lists to simplify checking for win conditions at the end. A few things we need to consider for this to work:
* Deciding on x or o for the player/computer
* Printing out the grid and how we will ask which square player wants to move to
* Making sure the player doesn't move anywhere where the computer already moved and vice versa
* Checking the computer and player moves against win conditions, and keep playing until satisfied (or grid is full)

````python
#tic-tac-toe two-player
import math
import random

class tictactoe:
    def __init__(self):
        self.dim = 3
        self.move_list = []
        self.player_list = []
        self.computer_list = []
        self.win_check = False
            
    def player(self):
        ask_player = input("Are you x or o? ")
        while ask_player.lower() != "x" and ask_player.lower() != "o":
            ask_player = input("Are you x or o? ")
        ask_player = ask_player.lower()
        print("You are: " + ask_player)
        self.ask_player = " " + ask_player + " "
        
        if self.ask_player == ' x ':
            self.other_player = ' o '
        else:
            self.other_player = ' x '
            
    
    def define_grid(self):
        minimum = 1
        maximum = self.dim * self.dim
        
        print('\n')
        
        row_list = []
        grid_list = []
        
        for row in range(self.dim):
            for column in range(1, self.dim+1):
                print(str(column + self.dim * row), end=' ')
            print('\n')
        
        print("Current grid: ")
        
        for row in range(self.dim):
            for column in range(1, self.dim+1):
                print(' * ', end=' ')
                row_list.append(' * ')
            print('\n')
            grid_list.append(row_list)
            row_list = []
        
        self.grid_list = grid_list
        
        print('\n')
        
    def print_grid(self):
        
        for row in range(len(self.grid_list)):
            for column in range(len(self.grid_list)):
                print(self.grid_list[row][column], end = ' ')
            print('\n')
        
    def player_move(self):
        ask_square = int(input('Where do you want to move? Number 1-9: '))
        while int(ask_square)<1 or int(ask_square)>9:
            ask_square =int(input('Must be between 1 to 9: '))
        
        row_position = math.floor((ask_square - 1) / self.dim)
        
        column_position = math.floor((ask_square - 1) % self.dim)
        
        #print('your row position is: ' + str(row_position))
        #print('your column position is: ' + str(column_position)) 

        self.grid_list[row_position][column_position] = self.ask_player
        
        self.move_list.append(ask_square)
        self.player_list.append(ask_square)
    
    def computer_move(self):
        
        square = random.randint(1,9)
        while square in self.move_list:
            square = random.randint(1,9)
        
        print("The computer moved to: " + str(square))
        
        self.move_list.append(square)
        self.computer_list.append(square)
        
        row_position = math.floor((square - 1) / self.dim)
        
        column_position = math.floor((square - 1) % self.dim)
         
        self.grid_list[row_position][column_position] = self.other_player
        
        #print(self.move_list)
        
    def check_wins(self):
        
        self.win_conditions = [
            {1,2,3},
            {4,5,6},
            {7,8,9},
            {1,5,9},
            {3,5,7},
            {1,4,7},
            {2,5,8},
            {3,6,9}
        ]
        
        #check wins for the computer 
        for i in range(len(self.win_conditions)):
            result_computer = self.win_conditions[i].intersection(self.computer_list)
            result_player = self.win_conditions[i].intersection(self.player_list) 

            
            if len(result_computer)==3:
                print('The computer won! Better luck next time')
                self.win_check = True
                
            elif len(result_player)==3:
                print('Congratulations! You won')
                self.win_check = True
            
        if self.win_check == False:
            print("No wins yet!")

        
        
````

````python
new = tictactoe()
new.player()

new.define_grid()

while new.win_check == False:
    new.player_move()
    new.print_grid()
    new.check_wins()
    new.computer_move()
    new.print_grid()
    new.check_wins()
````
